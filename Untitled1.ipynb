```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
edges = [(1,2), (1,3), (1,4), (2,5), (3,6), (4,7), (5,8)]
G.add_edges_from(edges)
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='lightblue')
nx.draw_networkx_edges(G, pos, width=2, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=12, font_family='sans-serif')
plt.title("Визуализация графа", fontsize=15)
plt.axis('off')
plt.show()

def create_a_matrix(edges, num_vert):
  a_matrix = [[0 for _ in range(num_vert)] for _ in range(num_vert)]
  for edge in edges:
    u, v = edge
    a_matrix[u-1][v-1] = 1
    a_matrix[v-1][u-1] = 1
  return a_matrix
edges = [(1,2), (1,3), (1,4), (2,5), (3,6), (4,7), (5,8)]
num_vert = 8
a_matrix = create_a_matrix(edges, num_vert)
for row in a_matrix:
  print(row)

print("*" * 50)

def create_i_matrix(edges, num_vert):
  num_edges = len(edges)
  i_matrix = [[0 for _ in range(num_edges)] for _ in range(num_vert)]
  for edge_index, edge in enumerate(edges):
    u, v = edge
    i_matrix[u-1][edge_index] = 1
    i_matrix[v-1][edge_index] = 1
  return i_matrix
i_matrix = create_i_matrix(edges, num_vert)
for row in i_matrix:
 print(row)

print("*" * 50)

def find_con_components(a_matrix):
    num_vert = len(a_matrix)
    visited = [False] * num_vert
    components = []
    def dfs(vert, curr_comp):
        visited[vert] = True
        curr_comp.append(vert + 1)
        for neighbor, is_connected in enumerate(a_matrix[vert]):
            if is_connected and not visited[neighbor]:
                dfs(neighbor, curr_comp)
    for vert in range(num_vert):
        if not visited[vert]:
            curr_comp = []
            dfs(vert, curr_comp)
            components.append(curr_comp)
    return components
connected_components = find_con_components(a_matrix)
print("Компоненты связности:", connected_components)
```
